```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a time array (1 Hz sampling frequency, 10 seconds duration)
T = 10  # Duration in seconds
f_s = 100  # Sampling frequency in Hz
t = np.linspace(0, T, T*f_s, endpoint=False)
N = len(t)  # Number of samples

# Generate a signal (sum of two sine waves)
f1, f2 = 4, 10  # Frequencies in Hz
a1, a2 = 1.5, 4
y = a1*np.sin(2*np.pi*f1*t) + a2*np.sin(2*np.pi*f2*t)

# Compute FFT and scale by the number of samples
y_fft = 1/N * np.fft.fft(y)
frequencies = np.fft.fftfreq(N, 1/f_s)

# Plot original signal
plt.subplot(2, 1, 1)
plt.title("Original Signal: 1.5 Hz + 4 Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.plot(t, y)
plt.xlim(0, 1)
plt.grid(True)  

# Plot FFT
plt.subplot(2, 1, 2)
plt.title("FFT of Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.stem(frequencies, np.abs(y_fft)*2)  # Use stem plot to plot discrete values, multiply by 2 to account for the mirrored half
plt.xlim(0, 20)
plt.grid(True)
plt.tight_layout()
plt.show()
```
